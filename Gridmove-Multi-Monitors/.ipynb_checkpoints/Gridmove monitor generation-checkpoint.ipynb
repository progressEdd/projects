* N = number of monitors

```

```

# Generating the grid

````python
currentDate = datetime.datetime.now()
date = currentDate.strftime("%c")
current_time = currentDate.strftime("%H-%M-%S")
fdate = currentDate.strftime("%b-%d-%Y")
````

````python
groups = 66
monitors = 3
monitor = 1
numberOfGroups = groups * monitors
counter = 0

header = F"""; grid: {monitors}Mxipergrid1
; generated: {date}"""

filename = F"""{monitors}M-xipergrid1-{fdate}-{current_time}.grid"""

with open(filename, 'wt') as f:
    print(header, file=f)
    print("", file=f)
    print("", file=f)
    print("[Groups]", file=f)
    print("", file=f)
    print("NumberOfGroups = " + str(numberOfGroups), file=f)
    print("", file=f)

    for group in range(1,groups*monitors+1):
        if (group%66==0):
            monitor +=1
        if(counter%66==0):
            counter=0
        # print(counter)
        print("["+str(group)+"]"+ split[counter].format(monitor=monitor), file=f)
        print("", file=f)
        print("", file=f)
        counter+=1
````
