```python
import pandas as pd
sheet_url = 'https://docs.google.com/spreadsheets/d/1CqlU5LHuFKxq_Yt4i1F1aZdtTp6YnrFePcy4TQUKufg/edit#gid=0'
csv_export_url = sheet_url.replace('/edit#gid=', '/export?format=csv&gid=')
df = pd.read_csv(csv_export_url)        
df
```

```python
# function that will do grouping and return data frame
def grouping(df_fun):
    df_fun["count"] = 1
    g = df_fun.groupby("map")
    df_g = g.agg("sum").reset_index()
    return df_g
```

```python
df['victories'] = df.apply(lambda row: 1 if row.result == 'Victory' else 0, axis = 1)
df['defeats'] = df.apply(lambda row: 1 if row.result == 'Defeat' else 0, axis = 1)
```

```python
df_match = df[(df["result"] == 'Victory') | (df["result"] == 'Defeat')]
df_match
```

```python
df_test = grouping(df_match)
```

```python
df_test
```

```python
df_test["average victories"] = df_test["victories"]/df_test["count"]
df_test["average defeats"] = df_test["defeats"]/df_test["count"]
```
